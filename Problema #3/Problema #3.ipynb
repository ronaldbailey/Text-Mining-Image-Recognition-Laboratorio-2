```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def visualizar_imagen_3D(ruta_imagen):

    # Leer la imagen en escala de grises
    imagen = cv2.imread(ruta_imagen, cv2.IMREAD_GRAYSCALE)

    # Crear una malla de coordenadas para la imagen
    x = np.arange(imagen.shape[1])
    y = np.arange(imagen.shape[0])
    x, y = np.meshgrid(x, y)

    # Crear la figura y el eje 3D
    fig = plt.figure(figsize=(10, 10))
    ax = fig.add_subplot(111, projection='3d')

    # Visualizar la superficie
    ax.plot_surface(x, y, imagen, cmap='gray')

    # Configurar los ejes
    ax.set_xlabel('Eje X')
    ax.set_ylabel('Eje Y')
    ax.set_zlabel('Intensidad')
    ax.set_title('Visualización 3D de la Imagen')

    plt.show()

# Ejemplo de uso
ruta = '../imagenes/carro.jpg'
visualizar_imagen_3D(ruta)
```
