```python
import cv2
import numpy as np
import matplotlib.pyplot as plt 

def binarizar_imagen(ruta_imagen, umbral=255):

    # Leer la imagen en escala de grises
    imagen = cv2.imread(ruta_imagen, cv2.IMREAD_GRAYSCALE)

    # Binarizar la imagen manualmente
    imagen_binarizada = np.where(imagen > umbral, 255, 0).astype(np.uint8)

    return imagen_binarizada

# Ejemplo de uso
ruta = '../imagenes/ifel.jpg'
imagen_bin = binarizar_imagen(ruta, 127)
plt.imshow(imagen_bin, cmap='gray')
```
